```python
import json
import pandas as pd
```

```python
data_train = pd.read_csv('data_train.csv')
len(data_train)
```

```python
concept_labels = json.load(open('concept_labels.json', 'r'))
len(concept_labels)
```

```python
data_train_labeled = data_train[data_train['image_file'].isin(concept_labels.keys())]
data_train_unlabeled = data_train[~data_train['image_file'].isin(concept_labels.keys())]
len(data_train_labeled), len(data_train_unlabeled)
```

```python
data_train_unlabeled = data_train_unlabeled.sample(frac=1, random_state=42).reset_index(drop=True)
split_size = len(data_train_unlabeled) // 4
data_train1 = data_train_unlabeled.iloc[:split_size]
data_train2 = data_train_unlabeled.iloc[split_size:2*split_size]
data_train3 = data_train_unlabeled.iloc[2*split_size:3*split_size]
data_train4 = data_train_unlabeled.iloc[3*split_size:]
print(len(data_train1), len(data_train2), len(data_train3), len(data_train4))
```

```python
data_train1 = pd.concat([data_train_labeled, data_train1]).reset_index(drop=True)
data_train2 = pd.concat([data_train_labeled, data_train2]).reset_index(drop=True)
data_train3 = pd.concat([data_train_labeled, data_train3]).reset_index(drop=True)
data_train4 = pd.concat([data_train_labeled, data_train4]).reset_index(drop=True)
print(len(data_train1), len(data_train2), len(data_train3), len(data_train4))
```

```python
data_train1.to_csv('data_train1.csv', index=False)
data_train2.to_csv('data_train2.csv', index=False)
data_train3.to_csv('data_train3.csv', index=False)
data_train4.to_csv('data_train4.csv', index=False)
```
