```python
import json
```

```python
merged = []

for k in range(1, 5):
    data = json.load(open(f'all_samples{k}.json'))
    images = dict()

    for i in range(len(data)):
        if data[i]['image'] not in images:
            images[data[i]['image']] = data[i]
        else:
            # print(f"Duplicate found: {data[i]['image']} in all_samples{k}.json")
            old_score = sum(list(images[data[i]['image']]['se_scores'].values()))
            new_score = sum(list(data[i]['se_scores'].values()))
            if new_score > old_score:
                images[data[i]['image']] = data[i]

    print(f"Added {len(images)} unique images from all_samples{k}.json")
    merged.extend(list(images.values()))

print(len(merged))
```

```python
with open('all_samples.json', 'w') as f:
    json.dump(merged, f, indent=2)
```
