```python
# Merge Alan and Nhat's results
import json

alan = json.load(open('checkpoints/cub200_experiment_llama_wo_human/classification/cub200/test_samples_evaluation_alan.json', 'r'))
nhat = json.load(open('checkpoints/cub200_experiment_llama_wo_human/classification/cub200/test_samples_evaluation_nhat.json', 'r'))
with open("data/cub200/test_set200.txt", "r") as f:
    image_files_names = f.read().splitlines()

def is_nhat_sample(key):
    if key[:4] == "100.":
        return True
    
    if key[0] == "1" or key[0] == "2":
        return False
    
    return True

merged = {}

for image in image_files_names:
    if is_nhat_sample(image):
        merged[image] = nhat[image]
    else:
        merged[image] = alan[image]

len(merged)
with open("checkpoints/cub200_experiment_llama_wo_human/classification/cub200/test_samples_evaluation_all.json", "w") as f:
    json.dump(merged, f, indent=2)
```

```python
import json

test_samples = json.load(open('checkpoints/cub200_experiment_llama_wo_human/classification/cub200/test_samples_evaluation_all.json', 'r'))
```

```python
total = 0
correct = 0
for sample, concepts in test_samples.items():
    for attr, score in concepts.items():
        total += 1
        correct += score
accuracy = correct / total if total > 0 else 0
print(f"Total samples: {total}")
print(f"Accuracy: {accuracy:.2%}")
```
