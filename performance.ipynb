```python
import json

test_samples = json.load(open('checkpoints/cub200_experiment_llama_core/classification/cub200/test_samples_evaluation_all.json', 'r'))
```

```python
total = 0
correct = 0
for sample, concepts in test_samples.items():
    for attr, score in concepts.items():
        total += 1
        correct += score
accuracy = correct / total if total > 0 else 0
print(f"Total samples: {total}")
print(f"Accuracy: {accuracy:.2%}")
```
